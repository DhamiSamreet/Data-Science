```
import pandas as pd
import csv
csv.field_size_limit(1000000000)
df = pd.read_csv("state-of-the-union.csv")
df.columns=['Year','speech']

df.head()
```

```
from gensim.utils import simple_preprocess
texts = df.speech.apply(simple_preprocess)
#print(texts)
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stoplist = stopwords.words('english')
stoplist.append(['government','states','united'])
for i in range(len(texts)):
  remove=[]
  for k in range(len(texts[i])):
    if stoplist.count(texts[i][k])!=0:
      remove.append(texts[i][k])
  for k in remove:
    texts[i].remove(k)

print(stoplist)
```

```
print(texts)
```

```
from gensim import corpora

dictionary = corpora.Dictionary(texts)

corpus = [dictionary.doc2bow(text) for text in texts]
```

```
print(dictionary)
```

```
from gensim import models

tfidf = models.TfidfModel(corpus)
corpus_tfidf = tfidf[corpus]
```

```
for doc in corpus_tfidf:
    print(doc)
```

```
n_topics = 15

# Build an LSI model
lsi_model = models.LsiModel(corpus_tfidf,id2word=dictionary,num_topics=n_topics)
```

```
lsi_model.print_topics()
```

```
lda_model = models.LdaModel(corpus_tfidf, id2word=dictionary, num_topics=n_topics)
```

```
lda_model.print_topics()
```

```
for idx, topic in lda_model.show_topics(formatted=False, num_words= 20):
    print('Topic: {} \nWords: {}'.format(idx, '|'.join([w[0] for w in topic])))
```

```
def show_topics(ldamodel=lda_model, corpus=corpus, texts=texts):
    # Init output
    sent_topics_df = pd.DataFrame()

    # Get main topic in each document
    for i, row in enumerate(ldamodel[corpus]):
        row = sorted(row, key=lambda x: (x[1]), reverse=True)
        # Get the Dominant topic, Perc Contribution and Keywords for each document
        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0:  # => dominant topic
                wp = ldamodel.show_topic(topic_num)
                topic_keywords = ", ".join([word for word, prop in wp])
                sent_topics_df = sent_topics_df.append(pd.Series([int(topic_num), round(prop_topic,4), topic_keywords]), ignore_index=True)
            else:
                break
    sent_topics_df.columns = ['Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords']

    # Add original text to the end of the output
    contents = pd.Series(texts)
    sent_topics_df = pd.concat([sent_topics_df, contents], axis=1)
    return(sent_topics_df)


df_topic_keywords = show_topics(ldamodel=lda_model, corpus=corpus, texts=texts)

# Format
df_dominant_topic = df_topic_keywords.reset_index()
df_dominant_topic.columns = ['Document_No', 'Dominant_Topic', 'Topic_Perc_Contrib', 'Keywords', 'Text']

# Show
df_dominant_topic.head(10)
```

```
ap_df = pd.read_csv("ap.csv")
ap_df.head()
```

```
ap_df.columns=['index_code','Story']
ap_df.head()
ap_df.shape
```

```
ap_df.Story = ap_df.Story.str.replace("[^A-Za-z ]", " ")
texts_ap = ap_df.Story.astype(str).apply(simple_preprocess)
```

```
dictionary = corpora.Dictionary(texts_ap)
dictionary.filter_extremes(no_below=5, no_above=0.5)

corpus = [dictionary.doc2bow(text) for text in texts_ap]

tfidf = models.TfidfModel(corpus)
ap_tfidf = tfidf[corpus]
n_topics = 100
```

```
lda_model = models.LdaModel(ap_tfidf, id2word=dictionary, num_topics=n_topics)
```

```
lda_model.print_topics()
```

```
topics = lda_model.show_topics(num_topics=n_topics,num_words=20)
```

```
for topic in topics:
  print(f"Topic-#{topic[0]}: {topic[1]}.....\n\n")
```
